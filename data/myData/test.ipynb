```python
import matplotlib.pyplot as plt
import pandas as pd

# 创建ID到名称的映射字典
category_map = {
    1: "Finance",
    2: "Anime",
    3: "Sports",
    4: "Drama",
    5: "Technology",
    6: "Travel",
    7: "Food",
    8: "Pets",
    9: "Emotion",
    10: "Automotive",
    11: "Humanities and Social Sciences",
    12: "Agriculture",
    13: "Home and Living",
    14: "Life Log",
    15: "Social and Political News",
    16: "Fashion",
    17: "Dance",
    18: "Campus Education",
    19: "Film and Television",
    20: "Music",
    21: "Medical and Health",
    22: "Gaming",
    23: "Knowledge and Science"
}

# 将category列中的数字ID替换为对应的名称
df["category"]=df['category'].apply(lambda x:x if x<24 and x>0 else -1)
df['category_name'] = df['category'].apply(lambda x: category_map.get(x, 'Unknown'))
```

```python
# 检查数据是否已经正确转换
print(df['category_name'].head())

# 绘制直方图
plt.figure(figsize=(12, 8))  # 设置图形的大小
plt.hist(df['category_name'], bins=len(category_map), color='skyblue', edgecolor='black')  # 绘制直方图
plt.xticks(rotation=90)  # 如果类别名称较长，可以旋转90度以便于阅读
plt.xlabel('Category')  # 设置x轴标签
plt.ylabel('Frequency')  # 设置y轴标签
plt.title('视频主题分布图')  # 设置标题
plt.show()  # 显示图形
```

```python
# 将 DataFrame 保存到 Excel 文件
df.to_excel('视频标签_19136.xlsx', index=False, engine='openpyxl')
```
