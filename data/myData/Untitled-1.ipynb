```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.linspace(-10, 10, 100)
sig = sigmoid(x)

# 计算梯度
grad_pos = sig * (1 - sig)
grad_neg = (1 - sig) * sig

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# 绘制正梯度图
axes[0].plot(x, grad_pos, label='σ(x)·(1 - σ(x))')
axes[0].axvline(x=0, color='gray', linestyle='--')
axes[0].axhline(y=0, color='gray', linestyle='--')
axes[0].set_title('Pos. Gradient $v_{u,i+}$')
axes[0].set_xlabel('$\hat{y}_u^+$')
axes[0].set_ylabel('$σ(x)·(1 - σ(x))$')
axes[0].fill_between(x, grad_pos, 0, where=(x > -5) & (x < 5), color='blue', alpha=0.3, label='noisy')

# 绘制负梯度图
axes[1].plot(x, grad_neg, label='(1 - σ(x))·σ(x)', color='orange')
axes[1].axvline(x=0, color='gray', linestyle='--')
axes[1].axhline(y=0, color='gray', linestyle='--')
axes[1].set_title('Neg. Gradient $v_{u,i-}$')
axes[1].set_xlabel('$\hat{y}_u^-$')
axes[1].set_ylabel('$(1 - σ(x))·σ(x)$')
axes[1].fill_between(x, grad_neg, 0, where=(x > -5) & (x < 5), color='red', alpha=0.3, label='unobserved')

axes[0].legend()
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']  # 指定默认字体为黑体
plt.rcParams['axes.unicode_minus'] = False  # 解决保存图像时负号'-'显示为方块的问题

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.linspace(-10, 10, 100)
sig = sigmoid(x)
one_minus_sig = 1 - sig

# 创建图形
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# 正梯度图 (左图)
axes[0].plot(x, one_minus_sig, label='1 - σ(x)', color='blue')
axes[0].axvline(x=0, color='gray', linestyle='--')  # 添加垂直线
axes[0].axhline(y=0, color='gray', linestyle='--')  # 添加水平线
axes[0].set_title('正梯度 $v_{u,i+}$')
axes[0].set_xlabel('$\hat{y}_u^+$')
axes[0].set_ylabel('$1 - σ(x)$')
axes[0].fill_between(x, one_minus_sig, 0, where=(x > -5) & (x < 5), color='blue', alpha=0.3, label='噪声(假阳性)')  # 高亮显示区域
axes[0].legend()

# 负梯度图 (右图)
axes[1].plot(x, -one_minus_sig, label='-(1 - σ(x))', color='orange')
axes[1].axvline(x=0, color='gray', linestyle='--')  # 添加垂直线
axes[1].axhline(y=0, color='gray', linestyle='--')  # 添加水平线
axes[1].set_title('负梯度 $v_{u,i-}$')
axes[1].set_xlabel('$\hat{y}_u^-$')
axes[1].set_ylabel('$-(1 - σ(x))$')
axes[1].fill_between(x, -one_minus_sig, 0, where=(x > -5) & (x < 5), color='red', alpha=0.3, label='未知(假阴性)')  # 高亮显示区域
axes[1].legend()

plt.tight_layout()
plt.show()
```
