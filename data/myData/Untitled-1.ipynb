```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.linspace(-10, 10, 100)
sig = sigmoid(x)

# 
grad_pos = sig * (1 - sig)
grad_neg = (1 - sig) * sig

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# 
axes[0].plot(x, grad_pos, label='(x)(1 - (x))')
axes[0].axvline(x=0, color='gray', linestyle='--')
axes[0].axhline(y=0, color='gray', linestyle='--')
axes[0].set_title('Pos. Gradient $v_{u,i+}$')
axes[0].set_xlabel('$\hat{y}_u^+$')
axes[0].set_ylabel('$(x)(1 - (x))$')
axes[0].fill_between(x, grad_pos, 0, where=(x > -5) & (x < 5), color='blue', alpha=0.3, label='noisy')

# 
axes[1].plot(x, grad_neg, label='(1 - (x))(x)', color='orange')
axes[1].axvline(x=0, color='gray', linestyle='--')
axes[1].axhline(y=0, color='gray', linestyle='--')
axes[1].set_title('Neg. Gradient $v_{u,i-}$')
axes[1].set_xlabel('$\hat{y}_u^-$')
axes[1].set_ylabel('$(1 - (x))(x)$')
axes[1].fill_between(x, grad_neg, 0, where=(x > -5) & (x < 5), color='red', alpha=0.3, label='unobserved')

axes[0].legend()
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']  # 
plt.rcParams['axes.unicode_minus'] = False  # '-'

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.linspace(-10, 10, 100)
sig = sigmoid(x)
one_minus_sig = 1 - sig

# 
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

#  ()
axes[0].plot(x, one_minus_sig, label='1 - (x)', color='blue')
axes[0].axvline(x=0, color='gray', linestyle='--')  # 
axes[0].axhline(y=0, color='gray', linestyle='--')  # 
axes[0].set_title(' $v_{u,i+}$')
axes[0].set_xlabel('$\hat{y}_u^+$')
axes[0].set_ylabel('$1 - (x)$')
axes[0].fill_between(x, one_minus_sig, 0, where=(x > -5) & (x < 5), color='blue', alpha=0.3, label='()')  # 
axes[0].legend()

#  ()
axes[1].plot(x, -one_minus_sig, label='-(1 - (x))', color='orange')
axes[1].axvline(x=0, color='gray', linestyle='--')  # 
axes[1].axhline(y=0, color='gray', linestyle='--')  # 
axes[1].set_title(' $v_{u,i-}$')
axes[1].set_xlabel('$\hat{y}_u^-$')
axes[1].set_ylabel('$-(1 - (x))$')
axes[1].fill_between(x, -one_minus_sig, 0, where=(x > -5) & (x < 5), color='red', alpha=0.3, label='()')  # 
axes[1].legend()

plt.tight_layout()
plt.show()
```
